```python
import requests
import bs4
```

```python
request = requests.get('https://www.moneycontrol.com/news/india/')
```

```python
print(type(request))
```

```python
print(request.text)
```

```python
soup = bs4.BeautifulSoup(request.text, 'html.parser')
```

```python
print(soup.title)
```

```python
para = soup.select('p')
```

```python
print(para)
```

```python
para_one = soup.select_one('p')
para_one_text = para_one.getText()
print(para_one_text)
```

```python
div = soup.find_all('li', attrs={'class': 'clearfix'})
print(div)
len(div)
```

```python
for attr in div:
    for anchor in div:
        print(anchor.find('img').get('title'))
        print(anchor.find('a').get('href'))
        print(anchor.find('img').get('src'))
        print('--------------------------------------------------')
```
